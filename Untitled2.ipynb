```python
cart=[]
cart.append("milk")
cart.append("chocolate")
cart.append("maida")
print(cart)
```

```python
cart=["books","pens"]
extension=["pencil"]
cart.extend(extension)
print(cart)
```

append adds one word to the list ,extend can add many words at the end of the list

```python
university=[]
rules=["uniform","id","shoes"]
for c in rules:
 university.append(c)
print(university)
```
